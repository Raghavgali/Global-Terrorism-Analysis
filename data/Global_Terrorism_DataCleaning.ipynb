### Renaming Columns 

```python
df2.rename(columns={'iyear': 'year', 'imonth':'month', 'iday':'day', 'country_txt': 'country', 'attacktype1_txt':'attacktype', 'targtype1_txt': 'targtype', 'targsubtype1_txt': 'targsubtype', 'gname': 'pgroup', 'weaptype1_txt': 'weaptype', 'weapsubtype1_txt': 'weapsubtype'}, inplace=True)
```

```python
df2.head()
```

### For some of the events the date of the actual event is unknown therefore just going ahead with month and the year column

```python
df2 = df2.drop(df2[df2['month']==0].index)
df2 = df2.drop(df2[df2['day']==0].index)
```

```python
position = 4
new_column = df2['year'].astype(str) + '/' + df2['month'].astype(str) + '/' + df2['day'].astype(str)
df2.insert(position, 'date', new_column)
```

```python
df2.head()
```

```python
df2['date'] = pd.to_datetime(df2['date'])
```
